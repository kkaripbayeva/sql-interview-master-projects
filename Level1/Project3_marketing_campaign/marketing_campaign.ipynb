```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load the data
df = pd.read_csv('data/marketing_campaign.csv', sep=';') # Adjust separator if needed

# 2. Calculate Total Spend
mnt_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Total_Spend'] = df[mnt_cols].sum(axis=1)
```

```python
# --- Plot 1: Campaign Acceptance Rates ---
campaigns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']
success_counts = df[campaigns].sum()

plt.figure(figsize=(10, 6))
sns.barplot(x=success_counts.index, y=success_counts.values, palette='viridis')
plt.title('Success Counts Across Marketing Campaigns', fontsize=14)
plt.ylabel('Number of Accepted Offers')
plt.show()
```

```python
# --- Plot 2: Income vs Total Spending (Regression) ---
plt.figure(figsize=(10, 6))
# Removing outliers for a clearer view (e.g., income > 200k)
filtered_df = df[df['Income'] < 200000]
sns.regplot(data=filtered_df, x='Income', y='Total_Spend', scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('Correlation: Annual Income vs Total Spending', fontsize=14)
plt.show()
```

```python
# --- Plot 3: Education Level and Spending ---
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Education', y='Total_Spend', palette='Set2')
plt.title('Spending Distribution by Education Level', fontsize=14)
plt.show()
```
